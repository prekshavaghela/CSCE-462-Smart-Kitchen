```python
import pandas as pd
import numpy as np
MINTEMP = 26.0

# high range of the sensor (this will be red on the screen)
MAXTEMP = 32.0

# how many color values we can have
COLORDEPTH = 1024
```

```python
#function to convert from pixel to C
def unmap_value(p, in_min, in_max, out_min, out_max):
    return (((p - out_min) * (in_max - in_min)) / (out_max - out_min)) + in_min
```

```python
revert = unmap_value(2046, MINTEMP, MAXTEMP, 0, COLORDEPTH - 1)
print(revert) 
```
